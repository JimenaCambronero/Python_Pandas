# Curso Pandas

### Que es Pandas

La biblioteca de software de código abierto Pandas está diseñada específicamente para la manipulación y el análisis de datos en el lenguaje Python. Es potente, flexible y fácil de usar.
Vas a poder:

**Manipular datos:** Te permite cargar datos desde diferentes fuentes, como archivos CSV, bases de datos incluso páginas web. Cargados los datos podes filtrar, ordenar, agregar o eliminar información.

**Limpieza de datos:** A menudo los datos que obtenemos pueden tener valores faltantes, errores tipográficos o información duplicada. Pandas ofrece herramientas para limpiar datos de manera sencilla.

**Análisis rápido:** Por ejemplo, calcular estadísticas como promedios, la mediana o la desviación estandar de tus datos.

**Visualización de datos:** Además de manipular datos, Pandas puede trabajr muy bien con bibliotecas de visualización de datos, crear gráficos y visualizaciones informativas.




### Estructura de Datos

Pandas tiene dos estructuras de datos principales:

**DataFrame:** Similar a una hoja de cálculo en Excel o una tabla en una base de datos, es una estructura bidimensional que organiza los datos en filas y columnas. Cada columna puede contener diferentes tipos de datos (números, texto, fechas, etc)

**Series:** Es una estructura unidimensional similar a una lista o un vector, pero con etiquetas en lugar de índices numéricos. Puedes pensar en una serie como una sola columna de un DataFrame.
