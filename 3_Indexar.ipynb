## Multiindice

El multiindice lo podemos asociar a las tablas dinámicas de Excel

```python
# Ejemplo de Datos de ventas
dicc = {
    'Cliente': ['Talleres Pepe','Talleres Pepe', 'Pinturas Jorge', 'Pinturas Jorge', 'Pinturas Jorge', 'Distribución Alba'],
    'Producto': ['Negro', 'Rojo', 'Negro', 'Rojo', 'Azul', 'Rojo'],
    'Cantidad': [1,2,1,3,4,5]
}

clientes = pd.DataFrame(dicc)
clientes
```

```python
df = pd.read_csv('/Users/jimenacambronero/Desktop/Proyectos para Portfolio/PythonPracticas/Datos/Frutos_Secos.csv', delimiter=';')
df.set_index(['Nombre Cliente','Tipo Producto'],inplace=True) #similar a tabla dinámica
df
```

```python
# Agrupamos según las ventas a los clientes
clientes.set_index(['Cliente','Producto'], inplace=True)
clientes
```

```python
# Si quiero saber todas las ventas de Talleres Pepe 
clientes.loc['Talleres Pepe']
```

```python
df.loc['Mas y Mas'] #.loc localiza el indice por nombre 
```

```python
# Ventas por producto de Talleres Pepe
clientes.loc['Talleres Pepe','Rojo']
```

```python
df.loc['Mas y Mas', 'Nueces']
```

Siempre hay que acceder por el índice principal, no puedes acceder unicamente por el secundario

Ejemplo:

    clientes.loc['rojo'] 

Esto nos dara error

```python
# Tenes que usar .xs() Cross Section
clientes.xs('Talleres Pepe')
```

```python
# Tenemos que conocer el nivel.
# Encontrar las cantidades vendidas de producto rojo 
clientes.xs('Rojo', level='Producto')
```

```python
# Podemos acceder igual que .loc
clientes.xs(('Talleres Pepe','Rojo'))
```

## Cambiar de Multiindice a Multicolumns

El método .unstack() en pandas se utiliza comúnmente en situaciones donde los datos están en un formato apilado o en un índice jerárquico (MultiIndex) y quieres descomponer una de las capas del índice jerárquico en columnas separadas para facilitar el análisis y la visualización de los datos.


```python
clientes
```

```python
clientes_columns = clientes.unstack(fill_value=0) #Le pongo fill_value=0 para que no salgan NaN
clientes_columns
```

```python
#Puedo cambiar el nivel que quiero de multicolumna
clientes_columns2 = clientes.unstack(level=-2,fill_value=0)
clientes_columns2
```

# De MultiColumns a Multiindice

```python
clientes_multindice = clientes_columns.stack()
clientes_multindice
```

```python
# Si lo quiero pasar solo a indice
clientes_columns2.stack().droplevel(-2)
```

# Como indexar con PANDAS

Hay una convención que es llamar 'df' al DataFrame que estamos analizando, ahora lo hemos estado llamando 'datos', pero lo vamos a pasar a 'df'

```python
df = pd.read_csv('/Users/jimenacambronero/Desktop/Proyectos para Portfolio/PythonPracticas/Datos/Frutos_Secos.csv', delimiter=';' ,index_col='id')
df
```

### Por Nombre

```python
# 1 columna
df.loc[:,'Nombre Cliente']
```

```python
# Varias volumnas, atento []. Las columnas se pasan como lista
df.loc[:,['Nombre Cliente','Fecha Venta']]
```

```python
# Rango columnas; cuando conocemos las columnas, acá no lo pasamos como lista
df.loc[:,'Nombre Cliente':'Fecha Venta']
```

### Por Posición

```python
df
```

```python
# Todos los registros de una columna
df.iloc[:,0]
```

```python
# Primera columna y segunda. 
df.iloc[:,[0,1]]
```

```python
# De la primera columna a la tercera
df.iloc[:,0:3]
```

### Por Criterio


```python
# Las columnas que contenga "Cliente" en su nombre
df.loc[:,df.columns.str.contains('Cliente')]
```

```python
#Que contenga la palabra "Cliente" y que tambien me ponga aquellas que empiezan por "Fecha"
df.loc[:,df.columns.str.contains('Cliente') | df.columns.str.startswith('Fecha')]
```

# Indexar por Filas

### Por Posición

```python
#Primera fila
df.iloc[[0]]
```

```python
#Fila 1 y 5, recordar doble corchete
df.iloc[[0,5]]
```

```python
#De la fila 1 a la 5 (rango)
df.iloc[0:5]
```

### Por Nombre

```python
df.set_index('Nombre Cliente',inplace=True)
df
```

```python
df.loc['Mercadona']
```

```python
df.loc[['Mercadona','Mas y Mas']]
```

### Por Criterio

```python
# Que cumpla una condición
df.loc[df['Cantidad Vendida (kg)']>1000]
```

```python
#Que cumpla varias condiciones
df.loc[(df['Cantidad Vendida (kg)'] > 1000) & (df['Tipo Producto'] == 'Almendras')]
```
